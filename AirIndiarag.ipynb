```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("short-stories-for-children-ingles-primaria-continuemos-estudiando.pdf")
```

```python
pip install pypdf
```

```python
document=loader.load()
context=document[0].page_content
```

```python
from langchain_core.prompts import PromptTemplate
prompt=PromptTemplate(input_variables=["context","question"],
                      template="""you are an intilligent assistant who helps to answer the question based on the context provided.
                      context:{context} 
                      question:{question}
                       your context shall be directly addressed to answer the question. if the answer is not found in the context, say you don't know.""")
```

```python
formatted_prompt = prompt.format(context=context, question="who is the writer of the book")
```

```python
result=chat.invoke(formatted_prompt)
```

```python
print(result.content)
```

Directory based bot

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders.pdf import PyPDFLoader

loader = DirectoryLoader(
    path="D:/Chatbot_AI/test",
    glob="**/*.pdf",
    loader_cls=PyPDFLoader
)

documents = loader.load()
```

```python
context = "\n".join([doc.page_content for doc in documents])
```

```python
formatted_prompt=prompt.format(context=context,question="how many projects are there under  UNDER E-KRANTI")
result=chat.invoke(formatted_prompt)
```

```python
print(result.content)
```
